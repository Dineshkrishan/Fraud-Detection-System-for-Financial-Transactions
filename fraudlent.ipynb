```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r'Synthetic_Financial_datasets_log.csv')

```

```python
df = dataset.drop(columns=['nameOrig', 'nameDest'], axis=1)
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder

# Assuming 'df' is your DataFrame and 'type' is the column causing the issue
le = LabelEncoder()
df['type'] = le.fit_transform(df['type'])
```

```python
x = df.iloc[:, :-1]
y = df['isFraud']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([ [10000, 5000, 170136.0, 160296.36,	0.0, 0.0, 0, 0]])
```

```python
import pickle

with open('fraudLR.pkl', 'wb') as f:
    pickle.dump(lr, f)

print("Model saved as fraudLR.pkl")
```

Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train, y_train)

```

```python
dtc.score(x_test, y_test)*100
```

```python

dtc.predict([ [10000, 5000, 170136.0, 160296.36,	0.0, 0.0, 0, 0]])
```

```python
import pickle

with open('fraudDTR.pkl', 'wb') as f:
    pickle.dump(dtc, f)

print("Model saved as fraudDTR.pkl")
```
